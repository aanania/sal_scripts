```python
import os
from lsst.obs.lsst.utils import readRawFile
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Read in image
dir = '/mnt/data/ats/mcm/'
atcamera_fname='AT-O-20180928-00015'
full_file_name=os.path.join(dir,atcamera_fname+'.fits')

dir = '/mnt/dmcs/2018-10-03'
atcamera_fname='AT-O-20181003-00099'
full_file_name=os.path.join(dir,atcamera_fname+'.fits')
```

```python
res = readRawFile(full_file_name)
```

```python
vmin = np.min(res.image.array)
vmax = np.mean(res.image.array)+3*np.std(res.image.array)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(res.image.array, cmap='gray', vmin=vmin, vmax=vmax)
```
